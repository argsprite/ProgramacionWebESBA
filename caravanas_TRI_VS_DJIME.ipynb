### Comparación de Contenido de Celdas (simulando 'CLOBs')

En un entorno de Colab/Python, 'CLOBs' (Character Large Objects) se manejarían como cadenas de texto (strings) que pueden ser muy largas. A continuación, definiremos dos variables de cadena en celdas separadas para simular tus 'celdas CLOB'.

```
# Celda 1: Definición de la primera 'CLOB'
clob_content_1 = """Este es el contenido de la primera celda. Puede ser un texto muy largo, como el contenido de un documento, un código fuente, o cualquier dato textual extenso. \nEste es un ejemplo de varias líneas para simular un CLOB."""

print("Contenido de CLOB 1 cargado.")
```

```
# Celda 2: Definición de la segunda 'CLOB'
clob_content_2 = """Este es el contenido de la segunda celda. Vamos a hacerla ligeramente diferente para ver la comparación. \nEste es un ejemplo de varias líneas para simular un CLOB."""

print("Contenido de CLOB 2 cargado.")
```

Ahora, vamos a comparar el contenido de ambas variables. Podemos hacer una comparación simple de igualdad o, para contenidos muy grandes y posibles diferencias sutiles, usar una biblioteca como `difflib` para obtener un resumen de las diferencias.

```
import difflib

# Comparación simple de igualdad
if clob_content_1 == clob_content_2:
    print("\nLas dos 'CLOBs' son idénticas.")
else:
    print("\nLas dos 'CLOBs' son diferentes.")

# Para una comparación más detallada, usando difflib
d = difflib.Differ()
diff = list(d.compare(clob_content_1.splitlines(), clob_content_2.splitlines()))

print("\n--- Diferencias detalladas (usando difflib) ---")
for line in diff:
    # Mostrar solo las líneas que son diferentes
    if line.startswith('+') or line.startswith('-') or line.startswith('?'):
        print(line)

print("---------------------------------------------")
```

### Comparación de Contenido de Columnas 'CARAVANA' y 'CARAVANARECIBIDAS'

Vamos a extraer los valores de las columnas 'CARAVANA' y 'CARAVANARECIBIDAS' del primer registro de `df_xls` y compararlos usando `difflib` para ver las diferencias detalladas.

```
# Extraer el contenido de las columnas 'CARAVANA' y 'CARAVANARECIBIDAS' del primer registro
caravana_content = str(df_xls.loc[0, 'CARAVANA'])
caravana_recibidas_content = str(df_xls.loc[0, 'CARAVANARECIBIDAS'])

print(f"Contenido de 'CARAVANA' (primer registro):\n{caravana_content[:100]}...") # Mostrar solo una parte si es muy largo
print(f"Contenido de 'CARAVANARECIBIDAS' (primer registro):\n{caravana_recibidas_content[:100]}...") # Mostrar solo una parte si es muy largo

# Comparación simple de igualdad
if caravana_content == caravana_recibidas_content:
    print("\nLas columnas 'CARAVANA' y 'CARAVANARECIBIDAS' son idénticas en este registro.")
else:
    print("\nLas columnas 'CARAVANA' y 'CARAVANARECIBIDAS' son diferentes en este registro.")

# Para una comparación más detallada, usando difflib
import difflib
d = difflib.Differ()
diff_columns = list(d.compare(caravana_content.splitlines(), caravana_recibidas_content.splitlines()))

print("\n--- Diferencias detalladas entre 'CARAVANA' y 'CARAVANARECIBIDAS' (usando difflib) ---")
for line in diff_columns:
    # Mostrar solo las líneas que son diferentes
    if line.startswith('+') or line.startswith('-') or line.startswith('?'):
        print(line)

print("----------------------------------------------------------------------------------")
```

### Traducción de la lógica Qlik Sense a Python (Pandas)

Vamos a replicar los pasos de tu script de Qlik Sense para procesar las columnas 'CARAVANA' y 'CARAVANARECIBIDAS' del `df_xls`.

```
# 1. Dividir las caravanas en filas individuales para CARAVANA
caravanas_split = df_xls.assign(CARAVANAS_EN_COLUMNA=df_xls['CARAVANA'].str.split(' ')).explode('CARAVANAS_EN_COLUMNA')
caravanas_split = caravanas_split[caravanas_split['CARAVANAS_EN_COLUMNA'] != ''] # Eliminar entradas vacías si las hay

print("Primeras caravanas divididas de la columna 'CARAVANA':")
display(caravanas_split[['ID', 'CARAVANAS_EN_COLUMNA']].head())
```

```
# 2. Dividir las caravanas en filas individuales para CARAVANARECIBIDAS
recibidas_split = df_xls.assign(RECIBIDAS_EN_COLUMNA=df_xls['CARAVANARECIBIDAS'].str.split(' ')).explode('RECIBIDAS_EN_COLUMNA')
recibidas_split = recibidas_split[recibidas_split['RECIBIDAS_EN_COLUMNA'] != ''] # Eliminar entradas vacías si las hay

print("Primeras caravanas divididas de la columna 'CARAVANARECIBIDAS':")
display(recibidas_split[['ID', 'RECIBIDAS_EN_COLUMNA']].head())
```

```
# 3. Crear un conjunto (set) de caravanas recibidas para una búsqueda eficiente (equivalente a MAP_RECIBIDAS)
map_recibidas_set = set(recibidas_split['RECIBIDAS_EN_COLUMNA'].unique())

print(f"Número total de caravanas únicas recibidas: {len(map_recibidas_set)}")
# print(list(map_recibidas_set)[:10]) # Descomentar para ver las primeras 10 caravanas recibidas
```

```
# 4. Identificar 'FALTANTES' y determinar el estado
# Creamos una copia para evitar SettingWithCopyWarning
faltantes_df = caravanas_split[['ID', 'CARAVANAS_EN_COLUMNA']].copy()
faltantes_df['estado'] = faltantes_df['CARAVANAS_EN_COLUMNA'].apply(lambda x: 'RECIBIDAS' if x in map_recibidas_set else 'NO RECIBIDAS')

print("Tabla resultante 'FALTANTES':")
display(faltantes_df.head(32))
```

### Identificación de 'REEMPLAZADAS'

Para identificar las caravanas 'REEMPLAZADAS', buscaremos aquellas que están presentes en la columna `CARAVANARECIBIDAS` pero no en la columna `CARAVANA` original. Esto implica crear un conjunto de las caravanas originales y luego filtrar las recibidas.

```
# Crear un conjunto (set) de caravanas originales para una búsqueda eficiente
map_caravanas_set = set(caravanas_split['CARAVANAS_EN_COLUMNA'].unique())

# Identificar 'REEMPLAZADAS'
# Creamos una copia para evitar SettingWithCopyWarning
reemplazadas_df = recibidas_split[['ID', 'RECIBIDAS_EN_COLUMNA']].copy()
reemplazadas_df['estado'] = reemplazadas_df['RECIBIDAS_EN_COLUMNA'].apply(lambda x: 'REEMPLAZADA' if x not in map_caravanas_set else 'NO REEMPLAZADA')

# Mostrar solo las caravanas que fueron 'REEMPLAZADAS'
reemplazadas_filtrado_df = reemplazadas_df[reemplazadas_df['estado'] == 'REEMPLAZADA']

print("Caravanas 'REEMPLAZADAS' identificadas:")
display(reemplazadas_filtrado_df.head())

print(f"Número total de caravanas 'REEMPLAZADAS': {len(reemplazadas_filtrado_df)}")
```

### Consolidación de 'NO RECIBIDAS' y 'REEMPLAZADAS'

Vamos a crear un único DataFrame que contenga tanto las caravanas identificadas como 'NO RECIBIDAS' como las 'REEMPLAZADAS'.

```
# Filtrar solo las caravanas 'NO RECIBIDAS'
no_recibidas_df = faltantes_df[faltantes_df['estado'] == 'NO RECIBIDAS'].copy()

# Renombrar la columna de caravanas en 'no_recibidas_df' para que coincida con 'reemplazadas_filtrado_df'
no_recibidas_df.rename(columns={'CARAVANAS_EN_COLUMNA': 'CARAVANA_AFECTADA'}, inplace=True)

# Renombrar la columna de caravanas en 'reemplazadas_filtrado_df'
reemplazadas_consolidado_df = reemplazadas_filtrado_df.rename(columns={'RECIBIDAS_EN_COLUMNA': 'CARAVANA_AFECTADA'}).copy()

# Combinar ambos DataFrames
df_consolidado_estados = pd.concat([no_recibidas_df, reemplazadas_consolidado_df], ignore_index=True)

print("DataFrame Consolidado de Estados (NO RECIBIDAS y REEMPLAZADAS):")
display(df_consolidado_estados)
```

### Mostrar el contexto original de las Caravanas Afectadas

Para ver las listas completas de 'CARAVANA' y 'CARAVANARECIBIDAS' para cada `ID` que tiene caravanas 'NO RECIBIDAS' o 'REEMPLAZADAS', fusionaremos el DataFrame `df_consolidado_estados` con el `df_xls` original.

```
# Fusionar df_consolidado_estados con df_xls para obtener las columnas CARAVANA y CARAVANARECIBIDAS
df_final_contexto = pd.merge(
    df_consolidado_estados,
    df_xls[['ID', 'CARAVANA', 'CARAVANARECIBIDAS']],
    on='ID',
    how='left'
)

print("DataFrame consolidado con el contexto original de 'CARAVANA' y 'CARAVANARECIBIDAS':")
display(df_final_contexto[['ID', 'CARAVANA_AFECTADA', 'estado', 'CARAVANA', 'CARAVANARECIBIDAS']])
```

### Consolidado de todas las Caravanas con sus Estados Detallados

Aquí se presenta un DataFrame con todas las caravanas únicas que aparecieron en las columnas 'CARAVANA' o 'CARAVANARECIBIDAS', junto con dos columnas de estado:
- **Estado_Desde_Envio**: Indica si la caravana, que fue parte de la lista original 'CARAVANA', fue 'RECIBIDAS' o 'NO RECIBIDAS'. Si no estaba en la lista original, se marcará como 'NO ENVIADA'.
- **Estado_Desde_Recepcion**: Indica si la caravana, que fue parte de la lista de 'CARAVANARECIBIDAS', fue 'REEMPLAZADA' o 'NO REEMPLAZADA'. Si no estaba en la lista de recibidas, se marcará como 'NO RECIBIDA EN CONTEO'.

```
# 1. Preparar el DataFrame de las caravanas esperadas/enviadas
df_estado_enviado = faltantes_df[['ID', 'CARAVANAS_EN_COLUMNA', 'estado']].copy()
df_estado_enviado.rename(columns={'CARAVANAS_EN_COLUMNA': 'Caravana_ID', 'estado': 'Estado_Desde_Envio'}, inplace=True)

# 2. Preparar el DataFrame de las caravanas realmente recibidas
df_estado_recibido = reemplazadas_df[['ID', 'RECIBIDAS_EN_COLUMNA', 'estado']].copy()
df_estado_recibido.rename(columns={'RECIBIDAS_EN_COLUMNA': 'Caravana_ID', 'estado': 'Estado_Desde_Recepcion'}, inplace=True)

# 3. Combinar ambos DataFrames para obtener un listado único de todas las caravanas y sus estados
# Usamos un 'outer' merge para incluir todas las caravanas que aparecieron en cualquiera de las dos listas
df_caravanas_estados_consolidado = pd.merge(
    df_estado_enviado,
    df_estado_recibido,
    on=['ID', 'Caravana_ID'],
    how='outer'
)

# 4. Rellenar los valores NaN para clarificar si una caravana no estaba en una de las listas originales
# Por ejemplo, si una caravana solo aparece en 'CARAVANARECIBIDAS' y no en 'CARAVANA', su 'Estado_Desde_Envio' será NaN.
# Si una caravana solo aparece en 'CARAVANA' y no en 'CARAVANARECIBIDAS', su 'Estado_Desde_Recepcion' será NaN.
df_caravanas_estados_consolidado['Estado_Desde_Envio'].fillna('NO ENVIADA', inplace=True)
df_caravanas_estados_consolidado['Estado_Desde_Recepcion'].fillna('NO RECIBIDA EN CONTEO', inplace=True)


print("Consolidado de todas las caravanas con sus dos estados:")
display(df_caravanas_estados_consolidado.sort_values(by=['ID', 'Caravana_ID']))
```

```
df_caravanas_estados_consolidado['Estado_Desde_Envio'] = df_caravanas_estados_consolidado['Estado_Desde_Envio'].replace({'RECIBIDAS': 'ENVIADAS', 'NO RECIBIDAS': 'NO ENVIADAS'})

print("Valores en 'Estado_Desde_Envio' actualizados:")
display(df_caravanas_estados_consolidado['Estado_Desde_Envio'].value_counts())
```
